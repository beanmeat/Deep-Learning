```python
import torch
import torch.nn as nn
```

```python
# 创建数据
x = torch.randn(3, 5)
t = torch.randn(3, 5)
print(x)
print(t)
```

```python
# 计算损失函数
loss = nn.L1Loss()
print(loss(x, t))
```

```python
# 计算损失函数
loss = nn.MSELoss()
print(loss(x, t))
```

```python
loss = nn.SmoothL1Loss()
print(loss(x, t))
```
