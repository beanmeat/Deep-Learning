```python
import torch
import torch.nn as nn
```

```python
# 1. 真实值为类别类型,一共六条数据，每条数据8个分类（0~7）
x = torch.randn(6,8)
print(x)
```

```python
t = torch.tensor([1,0,3,2,5,7])
print(t)
```

```python
# 计算损失
loss = nn.CrossEntropyLoss()
print(loss(x,t))
```

```python
# 真实值为分类概率或独热编码
t = torch.randn(6,8).softmax(dim=1)
print(t)
```

```python
# 计算损失
loss = nn.CrossEntropyLoss()
print(loss(x,t))
```

```python
input = torch.randn(6,8)
print(input)
target = torch.randint(low = 0, high = 8, size = (6,))
print(target)
```

```python
# 1. 直接调用CrossEntropyLoss
loss = nn.CrossEntropyLoss()
print(loss(input, target))
```

```python
# 2. 先调LogSoftmax，在调NLLLoss
loss1 = nn.LogSoftmax(dim=1)
loss2 = nn.NLLLoss()
y = loss1(input)
print(y)
print(loss2(y, target))
```
