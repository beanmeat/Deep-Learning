```python
import numpy as np
import torch
```

```python
# 1. 张量元素类型转换
tensor1 = torch.tensor([1,2,3])
print(tensor1, tensor1.dtype)
tensor1 = tensor1.type(torch.float32)
print(tensor1, tensor1.dtype)
```

```python
tensor1 = tensor1.long()
print(tensor1,tensor1.dtype)
```

```python
tensor1 = tensor1.bool()
print(tensor1,tensor1.dtype)
```

```python
# Tensor与ndarray转换
import torch

# 使用numpy()方法将Tensor转换为ndarray，共享内存
tensor1 = torch.rand(3, 2)
numpy_array = tensor1.numpy()
print(tensor1)
print(numpy_array)
print(type(tensor1), type(numpy_array))
print()
tensor1[:, 0] = 4
print(tensor1)
print(numpy_array)
print()

# 使用copy()方法避免共享内存
numpy_array = tensor1.numpy().copy()
tensor1[:, 0] = -1
print(tensor1)
print(numpy_array)
```

```python
import torch
import numpy as np

# 使用from_numpy()方法将ndarray转换为Tensor，共享内存
numpy_array = np.random.randn(3)
tensor1 = torch.from_numpy(numpy_array)
print(numpy_array)
print(tensor1)
print()
numpy_array[0] = 100
print(numpy_array)
print(tensor1)
print()

# 使用copy()方法避免共享内存
tensor1 = torch.from_numpy(numpy_array.copy())
numpy_array[0] = -1
print(numpy_array)
print(tensor1)
```

```python
import torch
import numpy as np

# 使用torch.tensor()将ndarray转换为Tensor
numpy_array = np.random.randn(3)
tensor1 = torch.tensor(numpy_array)
print(numpy_array)
print(tensor1)
print()
numpy_array[0] = 100
print(numpy_array)
print(tensor1)
```
