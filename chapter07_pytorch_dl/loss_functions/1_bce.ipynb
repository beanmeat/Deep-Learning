```python
import torch
import torch.nn as nn
```

```python
# 输入
x = torch.randn(3,2)
print(x)
```

```python
# 真实值（标签）
t = torch.tensor([[0,1],[1,0],[0,1]],dtype=torch.float)
print(t)
```

```python
# 预测值（输出）
y = torch.sigmoid(x)
print(y)
```

```python
# 定义损失函数，计算损失值
loss = nn.BCELoss()
print(loss(y,t))
```
