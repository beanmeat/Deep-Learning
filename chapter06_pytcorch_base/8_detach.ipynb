```python
import torch
```

```python
x = torch.tensor([1.0, 2, 3], requires_grad=True)
y = x.detach()

print(x)
print(y)
```

```python
print(id(x))
print(id(y))
```

```python
print(x.storage().data_ptr())
print(y.storage().data_ptr())
```

```python
z1 = x ** 2
print(z1)

z2 = y ** 2
print(z2)
```

```python
print(z1.requires_grad)
z1.sum().backward()
```

```python
print(z2.requires_grad)
z2.sum().backward()
```
