```python
import torch
```

```python
x1 = torch.tensor([1.,2.,3.,4.],requires_grad=True)
x2 = torch.tensor([1.,2.,3.,4.],requires_grad=True)
print(x1)
print(x2)
```

```python
y1 = x1.sigmoid()
y2 = x2.sigmoid()
print(y1)
print(y2)
```

```python
# 反向传播计算梯度
y1.sum().backward()
y2.sum().backward()
print("x1.grad = ",x1.grad)
print("x2.grad = ",x2.grad)
```

```python
z1 = y1.data
z2 = y2.detach()
print(z1)
print(z2)
print(z1.requires_grad)
print(z2.requires_grad)
print(y1.requires_grad)
print(y2.requires_grad)
```

```python
z1.zero_()
z2.zero_()
print(y1)
print(y2)
```
